````python
import os

import numpy as np
import pandas as pd
import scanpy as sc
import anndata
import matplotlib.pyplot as plt

import collections
%matplotlib inline

sc.settings.verbosity = 3             # verbosity: errors (0), warnings (1), info (2), hints (3)
sc.logging.print_versions()
````

````python
adata = anndata.read_h5ad(os.path.join("write", "tgfb1-2.h5ad"))
````

## How to increase DPI of plot.

Put

```
sc.set_figure_params(scanpy=True, dpi=600, dpi_save=600, frameon=True, vector_friendly=True, fontsize=10, color_map=None, format='pdf', transparent=False, ipython_format='png2x')
```

in a cell before the plots are printed.

To revert it back, use:

```
sc.set_figure_params(scanpy=True, dpi=80, dpi_save=150, frameon=True, vector_friendly=True, fontsize=10, color_map=None, format='pdf', transparent=False, ipython_format='png2x')
```

### Low DPI plot

````python
sc.tl.pca(adata, svd_solver='arpack')
````

````python
sc.pl.pca(adata, color="batch")
````

### High DPI plot

We use the above command in a cell to turn the high DPI mode on.

````python
sc.set_figure_params(scanpy=True, dpi=600, dpi_save=600, frameon=True, vector_friendly=True, fontsize=10, color_map=None, format='pdf', transparent=False, ipython_format='png2x')
````

Then, we print the same plot to make it high resolution.

````python
sc.pl.pca(adata, color="batch")
````

Then, we turn it back to low resolution.

````python
sc.set_figure_params(scanpy=True, dpi=80, dpi_save=150, frameon=True, vector_friendly=True, fontsize=10, color_map=None, format='pdf', transparent=False, ipython_format='png2x')
````
